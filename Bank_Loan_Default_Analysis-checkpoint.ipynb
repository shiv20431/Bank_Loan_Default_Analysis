```python
import pandas as pd
#1 Loading a data:
df = pd.read_csv(r"C:\Users\hp\OneDrive\Desktop\Loan_default.csv")
print(df)

# For Quick Look
print("SHAPES:\n",df.shape)
print("\nTOP 5 COLUMNS:\n",df.head(5))
```

```python
#1 CLEANING THE DATASET
# 1.Cleaning the column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ","_")
print(df)

# 2.Removing Duplicates Values
df.drop_duplicates()
```

```python
#2 Cheaking the missing values 
df.isnull().sum()
```

```python
# 3.Calculating Repayment Rate:
import pandas as pd
# Repayment Rate by Loan Amount

total_amount = df["loan_amount"].sum()
repaid_amount = df.loc[df["default"] == 0, "loan_amount"].sum()
repayment_rate_amount = repaid_amount / total_amount * 100

print(f"Repayment Rate (by amount): {repayment_rate_amount:.2f}%")
```

```python
#4 Corelating income, loan_amount and default
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Select relevant columns (adjust names if different in your dataset)
df_subset = df[['income', 'loan_amount', 'default']]

# Calculate correlation matrix
corr_matrix = df_subset.corr()

print("Correlation Matrix:\n", corr_matrix)

# Visualize correlation
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.show()
```

```python
#5 Segmenting Customers By Risk
# Risk score
df['risk_score'] = (df['loan_amount'] / df['income']) + (2 * df['default'])

# Segment customers
def segment(score):
    if score >= 3: return "High Risk"
    elif score >= 1.5: return "Medium Risk"
    else: return "Low Risk"

df['risk_segment'] = df['risk_score'].apply(segment)

print(df[['loan_id','income','loan_amount','default','risk_score','risk_segment']])
```

```python
#7 Loan approval trends

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

bins = [0, 600, 700, 750, 850]
labels = ['Poor', 'Fair', 'Good', 'Excellent']
df['credit_category'] = pd.cut(df['credit_score'], bins=bins, labels=labels)

approval_by_score = df.groupby('credit_category')['default'].value_counts(normalize=True).unstack()
print(approval_by_score)


sns.countplot(x="credit_category", hue="default", data=df)
plt.title("Loan Approval by Credit Score Category")
plt.show()

```

```python
#8. Top 10 Branches By Repayment Performance


# Calculate repayment performance (%)
total_loans = df["loan_amount"].sum()
repaid_loans = df.loc[df["default"] == 0, "loan_amount"].sum()

# Repayment performance (overall, not per branch)
repayment_performance = (repaid_loans / total_loans) * 100
print("Overall Repayment Performance:", repayment_performance)

perf = (
    df.groupby("loan_id")
      .apply(lambda x: (x.loc[x["default"] == 0, "loan_amount"].sum() / x["loan_amount"].sum()) * 100)
      .reset_index(name="Repayment_Performance")
)
# Top 10 branches
top10 = perf.sort_values(by="Repayment_Performance", ascending=False).head(10)
print(top10)



```
